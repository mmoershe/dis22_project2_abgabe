```python
import pandas as pd
import spacy
from collections import Counter

# Lade spaCy Modell
nlp = spacy.load("en_core_web_sm")

# Lade TSV mit pandas
df = pd.read_csv("cluster3.tsv", sep="\t")

# Hol dir die Queries als Liste
queries = df['query'].tolist()

# Fr POS-Muster
pos_patterns = []

for query in queries:
    doc = nlp(query)
    pos_patterns.append(" ".join([token.pos_ for token in doc]))

# Zhle die hufigsten POS-Muster
pos_counter = Counter(pos_patterns)

print("Top 10 POS-Muster:")
for pattern, count in pos_counter.most_common(10):
    print(f"{count}x: {pattern}")

# Fr Dependency-Muster (optional)
dep_patterns = []

for query in queries:
    doc = nlp(query)
    dep_patterns.append(" ".join([token.dep_ for token in doc]))

dep_counter = Counter(dep_patterns)

print("\nTop 10 Dependency-Muster:")
for pattern, count in dep_counter.most_common(10):
    print(f"{count}x: {pattern}")
```
